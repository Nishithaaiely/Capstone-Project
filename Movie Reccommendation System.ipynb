```python
#Importing Libraries

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Reading the Data

data = pd.read_csv("top10K-TMDB-movies.csv")
data.head()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
#Feature Selection

data = data[['id', 'title', 'overview', 'original_language', 'genre']]
data
```

```python
data = data.dropna().reset_index(inplace=False).drop(['index'], axis = 1)
data
```

```python
df = pd.DataFrame()
df['tags'] = data['overview']+data['genre']+data['original_language']
df['id'] = data['id']
df['title'] = data['title']
df
```

```python
#Count Vectorization

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=len(df), stop_words='english')
vectors = cv.fit_transform(df['tags'].values.astype('U')).toarray()
```

```python
#Cosine Similarity

from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
similarity.shape
```

```python
#Recommendation Logic

def recommend(df, movie):
    index = df[df['title'] == movie].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key = lambda x: x[1])
    for i in distance[0:10]:
        print(df.iloc[i[0]].title)
```

```python
recommend(df, 'Parasite')
```
